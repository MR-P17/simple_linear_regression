```
#import the require libraries
from math import sqrt
from csv import reader

#loading csv file to a proper list form dataset
def load_csv(filename):
  dataset=list()
  with open(filename, 'r') as file:
    csv_reader=reader(file)
    for row in csv_reader:
      if not row:
        continue
      dataset.append(row)
  return dataset
  
#converting string column to float
def str_col_to_float(dataset,col):
  length=len(dataset)
  dataset=dataset[1:length]
  clone_dataset=list()
  for i in range(len(dataset)):
    if len(dataset[i])==col:
      temp=list()
      for j in range(col):
        temp.append(float(dataset[i][j]))
      clone_dataset.append(temp)
  return clone_dataset
  

#loading train and test file
file1='/content/gdrive/My Drive/practice_ML_codes/train.csv'
tn_dataset=load_csv(file1)
file2='/content/gdrive/My Drive/practice_ML_codes/test.csv'
ts_dataset=load_csv(file2)
col1=len(tn_dataset[0])
train_dataset=str_col_to_float(tn_dataset,col1)
col2=len(ts_dataset[0])
test_dataset=str_col_to_float(ts_dataset,col2)


print(train_dataset)
print(test_dataset)

```

```
#calculation of mean
def mean(value):
  s=0.0
  for x in value:
    s+=x
  return s/float(len(value))

#calculation of variance
def var(values,mean):
  return sum([(x-mean)**2 for x in values])

#calculation of covariance
def cov(x,mean_x,y,mean_y):
  covar=0.0
  for i in range(len(x)):
    covar+=(x[i]-mean_x)*(y[i]-mean_y)
  return covar


#calculation of mean, variance and covariance of train set
x=[row[0] for row in train_dataset]
y=[row[1] for row in train_dataset]
mean_x , mean_y=mean(x),mean(y)
var_x,var_y=var(x,mean_x),var(y,mean_y)
covariance=cov(x,mean_x,y,mean_y)
print('mean_x=%.3f  variance_x=%.3f' %(mean_x,var_x))
print('mean_y=%.3f  variance_y=%.3f' %(mean_y,var_y))
print('covariance=%.3f' %(covariance))
```

```
#coefficient function
def coefficient(dataset):
  x=[row[0] for row in train_dataset]
  y=[row[1] for row in train_dataset]
  mean_x, mean_y=mean(x), mean(y)
  b1=cov(x,mean_x,y,mean_y)/var(x,mean_x)
  b0=mean_y-b1*mean_x
  return [b0,b1]


#calculation of coefficient b0 and b1
b0, b1=coefficient(train_dataset)
print('b0=%.3f  b1=%.3f'  %(b0,b1))
```

```
from random import randrange

# split function
def split_train_dataset(dataset,split):
  train=list()
  train_size=split*len(dataset)
  clone_dataset=list(dataset)
  while len(train)<train_size:
    index=randrange(len(clone_dataset))
    train.append(clone_dataset.pop(index))
  return [train,clone_dataset]

#claculation of rmse
def rmse_cal(actual,predicted):
  sum_error=0.0
  for i in range(len(actual)):
    pred_error=predicted[i]-actual[i]
    sum_error+=pred_error**2
  mean_error=sum_error/float(len(actual))
  return sqrt(mean_error)



#Algorithm for evaluation of rmse
def evaluate_algo(train_dataset,test_dataset,l_regression_algo,split):
  train,test=split_train_dataset(train_dataset,split)
  
  predicted_train=l_regression_algo(train,test)
  print(predicted_train)
  
  predicted_test=l_regression_algo(train,test_dataset)
  print(predicted_test)
  
  actual_train=[row[-1] for row in test]
  actual_test=[row[-1] for row in test_dataset]
  
  
  rmse_train=rmse_cal(actual_train,predicted_train)
  rmse_test=rmse_cal(actual_test,predicted_test)
  
  return [rmse_train,rmse_test]

#Linear_regre_algo
def l_regression_algo(train,test):
  predict=list()
  b0,b1=coefficient(train)
  for row in test:
    y1=b0+b1*row[0]
    predict.append(y1)
  return predict


#Test of linear regression
split=0.6
rmse_train,rmse_test=evaluate_algo(train_dataset,test_dataset,l_regression_algo,split)
print('RMSE_train: %.3f   RMSE_test: %.3f'  %(rmse_train,rmse_test))
```
