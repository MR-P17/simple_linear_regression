```
# Simple Linear Regression on the Swedish Insurance Dataset
from math import sqrt
from random import seed
from random import randrange
from csv import reader
from math import sqrt



# Load a CSV file
def load_csv(filename):
	dataset = list()
	with open(filename, 'r') as file:
		csv_reader = reader(file)
		for row in csv_reader:
			if not row:
				continue
			dataset.append(row)
	return dataset




# Convert string column to float of training and test set
def str_column_to_float(dataset, column):
  length=len(dataset)
  dataset=dataset[1:length]
  clone_dataset=list()
  for i in range(len(dataset)):
    if len(dataset[i])==column:
      temp=[]
      for j in range(column):
        temp.append(float(dataset[i][j]))
      clone_dataset.append(temp)
  return clone_dataset    
    
    
    
# load and preparing training and test data
filename = "/content/gdrive/My Drive/practice_ML_codes/train.csv"
test_file= "/content/gdrive/My Drive/practice_ML_codes/test.csv"
dataset = load_csv(filename)
test_set=load_csv(test_file)
column=len(dataset[0])
dataset_red=str_column_to_float(dataset, column)
column1=len(test_set[0])
test_data_red=str_column_to_float(test_set, column1)

  
  
  
```

```
# calculate mean
def mean(value):
  return sum(x)/float(len(value))

# calculate variance
def variance(value,mean):
  return sum([(x-mean)**2 for x in value])


# calculate covariance between x and y
def cov(x,mean_x,y,mean_y):
  covar=0.0
  for i in range(len(x)):
    covar+=(x[i]-mean_x)*(y[i]-mean_y)
  return covar

# calculate mean, variance and covariance 
x=[row[0] for row in dataset_red]
y=[row[1] for row in dataset_red]
mean_x, mean_y=mean(x),mean(y)
var_x, var_y=variance(x,mean_x),variance(y,mean_y)
covar=cov(x,mean_x,y,mean_y)
print('x stats: mean=%.3f  variance=%.3f' % (mean_x,var_x))
print('y stats: mean=%.3f  variance=%.3f'  %(mean_y,var_y))
print('covariance of x and y is:%.3f' %(covar))
```

```
#fuction for coefficient calculation
def coefficient(dataset):
  x=[row[0] for row in dataset]
  y=[row[1] for row in dataset]
  mean_x,mean_y=mean(x),mean(y)
  b1=cov(x,mean_x,y,mean_y)/variance(x,mean_x)
  b0=mean_y-b1*mean_x
  return [b0,b1]


#calculation of coefficient
b0,b1=coefficient(dataset_red)
print('b0=%.3f b1=%.3f' %(b0,b1))
```

```
# Split a dataset into a train and test set
def split_dataset(dataset,split):
  train=list()
  train_size=split*len(dataset)
  dataset_copy=list(dataset)
  while len(train)<train_size:
    index=randrange(len(dataset_copy))
    train.append(dataset_copy.pop(index))
  return train,dataset_copy

#calculation of rmse
def rmse_metric(actual,predicted):
  sum_error=0.0
  for i in range(len(actual)):
    prediction_error = predicted[i] - actual[i]
    sum_error += (prediction_error ** 2)
  mean_error = sum_error / float(len(actual))
  return sqrt(mean_error)


# Evaluate regression algorithm on training dataset
def evaluate_algo(dataset,test_set, algorithm,split,*args):
  train,test=split_dataset(dataset,split)
  
  
  predicted_train=algorithm(train,test,*args)
  print(predicted_train)
  predicted_test=algorithm(train,test_set,*args)
  print(predicted_test)
  
  
  actual_train=[row[-1] for row in test]
  actual_test=[row[-1] for row in test_set]
  
  rmse_train=rmse_metric(actual_train,predicted_train)
  rmse_test=rmse_metric(actual_test,predicted_test)
  
  return rmse_train,rmse_test



# defination of simple linear regression
def simple_linear_regression(train ,test):
  predict=list()
  b0,b1=coefficient(train)
  for row in test:
    y1=b0+b1*row[0]
    predict.append(y1)
  return predict
  
  


#Test simple linear regression
split=0.6
rmse_train,rmse_test=evaluate_algo(dataset_red,test_data_red,simple_linear_regression,split)
print('RMSE_train: %.3f' %(rmse_train))
print('RMSE_test: %.3f' %(rmse_test))
```
